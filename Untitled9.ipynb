```python
!pip install pyspark

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("SpotifyAnalytics").getOrCreate()

# Load the CSV file
df = spark.read.csv("/content/spotify-data.csv", header=True, inferSchema=True)
```

```python
from pyspark.sql.functions import col

# Convert important columns to proper types
numeric_cols = ["duration_ms", "year", "acousticness", "danceability", "energy",
                "instrumentalness", "liveness", "loudness", "speechiness",
                "tempo", "valence", "mode", "key", "popularity"]

for c in numeric_cols:
    df = df.withColumn(c, col(c).cast("double"))

df = df.withColumn("popularity", col("popularity").cast("int"))
```

```python
from pyspark.sql.functions import when

# Duration in minutes
df = df.withColumn("duration_min", col("duration_ms") / 60000)

# Popularity label
df = df.withColumn(
    "popularity_level",
    when(col("popularity") >= 80, "🔥 Super Hit")
    .when(col("popularity") >= 50, "👍 Hit")
    .otherwise("📉 Low")
)
df.show(10)
```

```python
df.groupBy("key").count().orderBy("count", ascending=False).show()
```

```python
df.groupBy("year").avg("popularity").orderBy("year").show()
```

```python
df.select("name","artists","popularity").orderBy("popularity",ascending=False).show(10)
```

```python
df.filter(df["explicit"] == 1).orderBy("valence", ascending=False).select("name", "artists", "valence","explicit").show(5)
```

```python
df.createOrReplaceTempView("songs")

spark.sql("""
      SELECT year,count(*) AS total_songs
      FROM songs
      GROUP BY year
      ORDER BY year
"""
).show()
```
